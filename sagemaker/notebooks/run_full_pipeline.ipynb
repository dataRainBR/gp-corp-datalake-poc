```python
#######################################################
# 1. IMPORTA OS MÓDULOS
#######################################################
import os
from sagemaker_dataset_builder import dataset_builder
from sagemaker_preprocessing import execute_preprocessing_pipeline
from sagemaker_preprocessing import (
    SalesAggregator,
    CalendarCompleter,
    NullHandler,
    TemporalFeatureCreator,
    LagFeatureCreator,
    RollingFeatureCreator,
    PriceFeatureCreator,
    BusinessFeatureCreator,
    TrendFeatureCreator,
    CategoryEncoder
)
from sagemaker_training import training_models
from sagemaker_inference import run_inference_pipeline
#######################################################
# 2. DEFINE VARIÁVEIS DE AMBIENTE
#######################################################
_BASE_DIR = os.path.dirname(os.path.abspath("__file_"))

DATASET_OUTPUT_PATH = os.path.join(_BASE_DIR, "dataset")
os.makedirs(DATASET_OUTPUT_PATH , exist_ok=True)

MODEL_OUTPUT_PATH=os.path.join(_BASE_DIR, "model")
os.makedirs(MODEL_OUTPUT_PATH , exist_ok=True)

OUTPUT_DIR = os.path.join(_BASE_DIR, "output")
os.makedirs(OUTPUT_DIR, exist_ok=True)


DATASET_IMPUT_PATH = os.path.join(DATASET_OUTPUT_PATH, "dataset_serie_temporal.parquet")
TRAIN_IMPUT_PATH = os.path.join(DATASET_OUTPUT_PATH, "dataset_train.parquet")
TEST_IMPUT_PATH = os.path.join(DATASET_OUTPUT_PATH, "dataset_test.parquet")
VAL_IMPUT_PATH = os.path.join(DATASET_OUTPUT_PATH, "dataset_val.parquet")


#######################################################
# 3. CARREGA BASE DE DADOS HISTORICA DO ATHENA
#######################################################
# dataset_builder(year_start=2024, year_end=2026, month_start=1, month_end=12, output_path=DATASET_OUTPUT_PATH)

#######################################################
# 4. REALIZA O PREPROCESSAMENTO DOS DADOS
#######################################################
preprocessing_pipeline = execute_preprocessing_pipeline(input_path=DATASET_IMPUT_PATH, output_path=DATASET_OUTPUT_PATH, train_ratio=0.7,val_ratio=0.15, test_ratio=0.15, output_format="parquet")

#######################################################
# 5. REALIZA O TREINAMENTO DO MODELO
#######################################################
training_models(train_path=TRAIN_IMPUT_PATH, test_path=TEST_IMPUT_PATH,  val_path=VAL_IMPUT_PATH, preprocessing_pipeline=preprocessing_pipeline, model_path=MODEL_OUTPUT_PATH)

#######################################################
# 6. REALIZA A PREDIÇÃO DO MODELO 
#######################################################
run_inference_pipeline()
```
